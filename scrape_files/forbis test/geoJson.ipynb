# Re-run this cell below when you need base dataframe

```python
x = x.rename(columns={"geometry_y":"geometry"})
x
```

```python
master_geojson = x
```

```python
master_geojson["Population"] = master_geojson["Population"].str.replace(".0", "")
master_geojson["Population"] = master_geojson["Population"].astype(int)
master_geojson
```

```python
master_geojson["Median Age"] = master_geojson["Median Age"].str.split(".")
master_geojson
```

```python
master_geojson["Median Age"] = master_geojson["Median Age"].str.get(0)

master_geojson
```

```python
master_geojson["Household Income"] = master_geojson["Household Income"].str.replace(".0", "")
master_geojson["Per Capita Income"] = master_geojson["Per Capita Income"].str.replace(".0", "")
master_geojson["Poverty Count"] = master_geojson["Poverty Count"].str.replace(".0", "")
master_geojson
```

```python
master_geojson["Population"] = master_geojson["Population"].astype(int)
```
