```python
import pandas as pd
import json
from geojson import MultiPoint
```

```python
geodata = pd.read_json("../../resources/data/testdata.json", orient='index', typ='series')
```

```python

# one set of properties for the first state (Maine)
# properties = geodata.features[0]["properties"]
# properties
geodata.features[0]["properties"]["NAME"]
# print(len(geodata.features))
all_states = []

for x in geodata.features:
    all_states.append(x["properties"]["NAME"])
    # print(x["properties"]["NAME"])
print(len(all_states))
all_states
```

```python
geodata.features[0]["geometry"]
```

```python
state_coords = []
# geodata.features[0]["geometry"]["coordinates"]
for x in geodata.features:
    state_coords.append(x["geometry"])
print(len(state_coords))
```

```python
coords_dict = {"state":all_states, "coordinates":state_coords}
coords_df = pd.DataFrame(coords_dict)
coords_df
```

```python
coords_df = coords_df.sort_values(by=["state"])
coords_df
```

```python
coords_df = coords_df.reset_index(drop=True)
print(len(coords_df["state"]))
coords_df
```

```python
coords_df_clean = coords_df.drop([8,39])
```

```python
coords_df_clean = coords_df_clean.reset_index(drop=True)
coords_df_clean
```

```python
print(len(coords_df_clean))
```

```python
coords_df_clean = coords_df_clean.rename(columns={"state":"Name"})
coords_df_clean.head()
```

```python
statedata = pd.read_csv("../../resources/data/alldata.csv")
all_states_sorted = coords_df_clean["Name"]
# print(len(all_states_sorted))
# all_states_sorted
statedata["Name"] = all_states_sorted
# print(len(statedata))
```

```python
statedata = pd.merge(statedata, coords_df_clean, on="Name")
statedata
```

```python
# statedata = statedata.rename(columns={"Name":"state"})
# statedata

# dropping state_coords column
statedata = statedata.drop(columns=["state_coords"])
statedata.head()
```

```python
# dropping extra coordinates column
statedata = statedata.drop(columns=["coordinates_y"])
statedata.head()
```

```python
# renaming coordinates_x column
statedata = statedata.rename(columns={"coordinates_x":"coordinates"})
statedata.head()
```

```python
# renaming coordinates_x column
statedata = statedata.rename(columns={"Name":"state"})
statedata.head()
```
