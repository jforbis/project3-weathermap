```python
import pandas as pd
import json
```

```python
statedata = pd.read_csv("../../resources/data/alldata.csv")
statedata
```

```python
master_dict = {}
states = list(statedata["Name"])
cities = list(statedata["Capital"])
lats = list(statedata["Lat"])
lngs = list(statedata["Lng"])
sts = list(statedata["State"])
pops = list(statedata["Population"])
med_ages = list(statedata["Median Age"])
hh_incomes = list(statedata["Household Income"])
pcis = list(statedata["Per Capita Income"])
pov_counts = list(statedata["Poverty Count"])
pov_rates = list(statedata["Poverty Rate"])

# url1 = "https://en.wikipedia.org/wiki/List_of_capitals_in_the_United_States"
```

```python
for i in range(50): #0,(len(states))):
    state=states[i]
    state = state.replace(" ", "-").lower()
    
    city=cities[i]
    city = city.replace(" ", "_").lower()
    city = city.replace("st.", "saint")
    
    lat=lats[i]
    lng=lngs[i]
    st=sts[i]
    pop=pops[i]
    med_age=med_ages[i]
    hh_income=hh_incomes[i]
    pci=pcis[i]
    pov_count=pov_counts[i]
    pov_rate=pov_rates[i]

    url="https://www.usclimatedata.com/climate/"+city+"/"+state+"/united-states/"
    
    response=pd.read_html(url)
    df0=response[0]
    df1=response[1]
    dict_1 = {
        "average high" : {
            "jan" : df0.iloc[0,1],
            "feb" : df0.iloc[0,2],
            "mar" : df0.iloc[0,3],
            "apr" : df0.iloc[0,4],
            "may" : df0.iloc[0,5],
            "jun" : df0.iloc[0,6],
            "jul" : df1.iloc[0,1],
            "aug" : df1.iloc[0,2],
            "sep" : df1.iloc[0,3],
            "oct" : df1.iloc[0,4],
            "nov" : df1.iloc[0,5],
            "dec" : df1.iloc[0,6]
        },
        "average low" : {
            "jan" : df0.iloc[1,1],
            "feb" : df0.iloc[1,2],
            "mar" : df0.iloc[1,3],
            "apr" : df0.iloc[1,4],
            "may" : df0.iloc[1,5],
            "jun" : df0.iloc[1,6],
            "jul" : df1.iloc[1,1],
            "aug" : df1.iloc[1,2],
            "sep" : df1.iloc[1,3],
            "oct" : df1.iloc[1,4],
            "nov" : df1.iloc[1,5],
            "dec" : df1.iloc[1,6]
         },
        "average rain" : {
            "jan" : df0.iloc[2,1],
            "feb" : df0.iloc[2,2],
            "mar" : df0.iloc[2,3],
            "apr" : df0.iloc[2,4],
            "may" : df0.iloc[2,5],
            "jun" : df0.iloc[2,6],
            "jul" : df1.iloc[2,1],
            "aug" : df1.iloc[2,2],
            "sep" : df1.iloc[2,3],
            "oct" : df1.iloc[2,4],
            "nov" : df1.iloc[2,5],
            "dec" : df1.iloc[2,6]
        },
        "Capital":city,
        "Lat":lat,
        "Lng":lng,
        "State":st,
        "Population":pop,
        "Median Age":med_age,
        "Household Income":hh_income,
        "Per Capita Income":pci,
        "Poverty Count":pov_count,
        "Poverty Rate":pov_rate
        }
    
    master_dict[state] = dict_1
print(url)
print()
print(master_dict)
```
